```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Impor the OpenWeatherMap API key
from api_keys import weather_api_key

# Import citipy to determine the cities based on latitude and longitude
from citipy import citipy
```

### Generate the Cities List by Using the `citipy` Library

```python
# Empty list for holding the latitude and longitude combinations
lat_lngs = []

# Empty list for holding the cities names
cities = []

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name

    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(f"Number of cities in the list: {len(cities)}")
```

## Requirement 2: Compute Linear Regression for Each Relationship


```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress

# Define a function to create Linear Regression plots

def plot_linear_regression(x, y, x_label, y_label, title, output_file):
    """
    Create a scatter plot with a linear regression line.
    
    Parameters:
    x (pd.Series): Independent variable data
    y (pd.Series): Dependent variable data
    x_label (str): Label for the x-axis
    y_label (str): Label for the y-axis
    title (str): Title of the plot
    output_file (str): File path to save the plot
    """
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    
    # Create scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='blue', edgecolor='black', s=100, alpha=0.7, label='Data points')
    
    # Plot the regression line
    regression_line = slope * x + intercept
    plt.plot(x, regression_line, color='red', label=f'Linear Regression\n$R^2={r_value**2:.2f}$')
    
    # Add titles and labels
    plt.title(title, fontsize=15)
    plt.xlabel(x_label, fontsize=12)
    plt.ylabel(y_label, fontsize=12)
    
    # Add gridlines
    plt.grid(True)
    
    # Add a legend
    plt.legend(loc='best')
    
    # Save the figure
    plt.savefig(output_file)
    
    # Show the plot
    plt.show()

# Data for demonstration
data = {
    'Latitude': [-22.8375, 42.6001, -21.3393, -42.7826, 59.7035, 42.097, 36.0564],
    'Temperature': [20.36, 10.36, 22.82, 11.72, 8.94, 5.77, 14.53]
}
df = pd.DataFrame(data)

# Call the function to create the plot
plot_linear_regression(df['Latitude'], df['Temperature'], 'Latitude', 'Temperature (Celsius)', 'Latitude vs. Temperature Linear Regression', 'output_data/latitude_vs_temperature_regression.png')
```

###  Temperature vs. Latitude Linear Regression Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Linear regression on Northern Hemisphere

# Data for demonstration 
data = {
    'City': ['Colorado', 'Albany', 'Saint-Pierre', 'New Norfolk', 'Nikolskoye', 'Jamestown', 'Lanzhou'],
    'Latitude': [-22.8375, 42.6001, -21.3393, -42.7826, 59.7035, 42.097, 36.0564],
    'Temperature': [20.36, 10.36, 22.82, 11.72, 8.94, 5.77, 14.53]
}

# Create the DataFrame
city_data_df = pd.DataFrame(data)

# Filter the data for the Northern Hemisphere (Latitude >= 0)
northern_hemisphere_df = city_data_df[city_data_df['Latitude'] >= 0]

# Function to create Linear Regression plots
def plot_linear_regression(x, y, x_label, y_label, title, output_file):
    """
    Create a scatter plot with a linear regression line.
    
    Parameters:
    x (pd.Series): Independent variable data
    y (pd.Series): Dependent variable data
    x_label (str): Label for the x-axis
    y_label (str): Label for the y-axis
    title (str): Title of the plot
    output_file (str): File path to save the plot
    """
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    
    # Create scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='blue', edgecolor='black', s=100, alpha=0.7, label='Data points')
    
    # Plot the regression line
    regression_line = slope * x + intercept
    plt.plot(x, regression_line, color='red', label=f'Linear Regression\n$R^2={r_value**2:.2f}$')
    
    # Add titles and labels
    plt.title(title, fontsize=15)
    plt.xlabel(x_label, fontsize=12)
    plt.ylabel(y_label, fontsize=12)
    
    # Add gridlines
    plt.grid(True)
    
    # Add a legend
    plt.legend(loc='best')
    
    # Save the figure
    plt.savefig(output_file)
    
    # Show the plot
    plt.show()

# Perform Linear Regression on Northern Hemisphere data
plot_linear_regression(
    northern_hemisphere_df['Latitude'], 
    northern_hemisphere_df['Temperature'], 
    'Latitude', 
    'Temperature (Celsius)', 
    'Linear Regression on Northern Hemisphere: Latitude vs. Temperature', 
    'output_data/northern_hemisphere_latitude_vs_temperature.png'
)
```
